# Basic Practice Numpy and Tensor

```python
import numpy as np
from pprint import pprint
```

Create arrays

```python
vector = np.array([1, 2, 3, 4, 5])
vector_slide = vector[2:]
pprint(vector)
pprint(vector_slide)
```

Slide is not a new object it is a view of vector so if modified a slided vector lead to modify original vector

```python
vector_slide[0] = 100
pprint(f"this is modifed original vector: {vector}, this is modified slide: {vector_slide}")
```

Array have attributes

```python
pprint(vector)
pprint(vector.ndim)
pprint(vector.size)
pprint(vector.shape)
pprint(vector.dtype)
```

The order of array shape is (batch size, row, col)

```python
pprint(f"this is matrice 2x3 of zeros: {np.zeros((2, 3))}") # create a 2x3 matrix of zeros
pprint(f"this is matrice 2x3 of ones: {np.ones((2, 3))}") # create a 2x3 matrix of ones
pprint(f"this is matrice 2x3 of empty values: {np.empty((2, 3))}") # create a 2x3 matrix of empty values, the values are random
```

```python
array1 = np.arange(0, 10, 2) # create an array from 0 to 10 with step 2
array2 = np.arange(10, 20, 2) # create an array from 0 to 10 with 6 elements
array2D = np.vstack((array1, array2)) # create a 2D array from 2 arrays
pprint(array2D)
```

# Basic Practice Tensor

```python
import tensorflow as tf
# create a scaler tensor
tf_scalar = tf.constant(1)
rank_1_tensor = tf.constant([1, 2, 3, 4, 5])
rank_2_tensor = tf.zeros([3, 2])
rank_3_tensor = tf.zeros([3, 2, 4])
rank_4_tensor = tf.zeros([3, 2, 4, 5])
print(tf_scalar)
pprint(f"type of every element: {rank_3_tensor.dtype}")
pprint(f"Number of axes: {rank_3_tensor.ndim}")
pprint(f"Shape of tensor: {rank_3_tensor.shape}")
pprint(f"number of elements a long axis 0: {rank_3_tensor.shape[0]}")
print(f"total elements in all axes: {tf.size(rank_3_tensor).numpy()}")
```

```python
x = tf.constant([[1], [2], [3]])
y  = tf.constant([1, 2, 3])
print(f"this is x: {x}")
print(f"this is x shape: {x.shape}")
print(f"this is y: {y.shape}")
```
